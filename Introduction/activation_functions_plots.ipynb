```python
import matplotlib.pyplot as plt
import numpy as np
```

## Sigmoid and tanh Activation Function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
def tanh(z):
    return np.tanh(z)
a = np.linspace(-np.pi,np.pi,1001)   
plt.plot(a, sigmoid(a),'b',label='sigmoid')
plt.plot(a, tanh(a),'r',label='tanh')
plt.legend()
plt.show()
```

## Relu Activation Function

```python
def relu(z):
    return max(0,z)
plt.plot(a, [relu(i) for i in a],'k',label='RelU')
plt.legend()
plt.show()
```
